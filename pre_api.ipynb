```python
# 로우데이터 검색후 저장
import pandas as pd
import requests
import json
import time
import datetime
import os
from pprint import pprint

# 나중에 주석처리
input_name = "아이폰 x"


# 함수시작
def search(input_name):  # 검색어 입력받음
    columns = [
        "category_id",
        "name",
        "product_image",
        "pid",
        "price",
        "update_time",
        "location",
    ]

    # 검색시작
    try:
        url = f"https://api.bunjang.co.kr/api/1/find_v2.json?q={input_name}&order=score&page=1&request_id=2024704081724&f_category_id=600700001&stat_device=w&n=100&stat_category_required=1&req_ref=search&version=5"
        # Request API
        response = requests.get(url)
        data = response.json()

        ipdf = pd.DataFrame(data["list"])[columns]
    except:
        return "검색어 입력오류"

    # nan값 제거
    ipdf["update_time"].sort_values(ascending=False)
    ipdf = ipdf[~(ipdf["location"] == "")]

    return ipdf.head()


# 결과 확인용
df = search(input_name)
df
```

```python
from apps.app import db
from apps.crud.models import User
from werkzeug.security import generate_password_hash

# 기존 사용자의 비밀번호를 업데이트하는 스크립트
def update_password_hash():
    users = User.query.all()
    for user in users:
        # 기존 비밀번호를 사용할 수 없기 때문에 새 비밀번호를 설정해야 합니다.
        # 예시로 모든 사용자 비밀번호를 'new_password'로 설정합니다.
        new_password = 'new_password'
        user.password_hash = generate_password_hash(new_password, method='pbkdf2:sha256')
        db.session.add(user)
    db.session.commit()

if __name__ == "__main__":
    update_password_hash()
    print("All user passwords have been updated.")
```
